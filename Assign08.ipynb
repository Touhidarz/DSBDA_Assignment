```python
print('Null in Age Before: ',df['Age'].isnull().sum())
print('Null in Survived : ',df['Survived'].isnull().sum())
print('Null in Fare : ',df['Fare'].isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
print('Null in Age After: ',df['Age'].isnull().sum())
```

```python
plt.hist(df['Age'])
plt.title('Age of Passenger')
plt.xlabel('Age')
plt.ylabel('count')
```

```python
plt.hist(df['Fare'])
plt.title('Fare of Ticket')
plt.xlabel('Fare')
plt.ylabel('Amount')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
# plt.hist()
axes[0].hist(df['Age'])
axes[1].hist(df['Sex'])
```
